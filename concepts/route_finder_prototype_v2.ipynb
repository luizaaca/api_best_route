```python
from osmnx_utils import initialize_graph, plot_route_segments_info
from route_plot_utils import RouteCalculator as RC
import random
```

```python
origin = "Praça da Sé, São Paulo"
destinations = [
    ("Edifício Copan, São Paulo", 1),
    ("Mercado Municipal de São Paulo", 2),
    ((-23.5465, -46.6367), 3),
    ((-23.561543, -46.656197), 2),
    ((-23.544534, -46.665030), 1),
    ((-23.534346, -46.640460), 3),
    ((-23.532278, -46.657468), 2),
    ((-23.531978, -46.634383), 1),
]
```

```python
g_proj, route_nodes = initialize_graph(origin, destinations)
```

```python
random_nodes = route_nodes[:1] + random.sample(route_nodes[1:], len(route_nodes) - 1)
print(random_nodes)
```

```python
route_calculator = RC(g_proj)
weight_function = route_calculator.get_weight_function()
```

```python
segments_info = route_calculator.compute_route_segments_info(random_nodes, weight_function, "priority")
```

```python
plot_route_segments_info(g_proj, segments_info)
```
